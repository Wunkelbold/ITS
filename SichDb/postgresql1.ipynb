```python
import psycopg2
import pandas as pd


try:
    conn = psycopg2.connect(host='192.168.56.101', database='postgres',user='postgres', password='postgres', port=5432)
except:
    print("Connecting to the database failed!")

cur = conn.cursor()

try:
    cur.execute("select * from airport")
except:
    print("Query execution failed!")

try:
    res = cur.fetchall()
except:
    print("Something went wrong")

 # close database cursor and database connection


for row in res:
    print(row[3])
    
cur.close()
conn.close()
```
