```python
#Aufgabe 7.0 a)
class Stack:                #Values of the kDict are stored as stack. Knots are stroed in value. Pop can be used to add connected knots the the keys.
    def __init__(self):
        self.st=[]
    
    def __repr__(self):
        return self.st

    def __push__(self,assBefehl):
        self.st.append(assBefehl)

    def __pop__(self):
        oberste=self.st[-1]
        self.st=self.st[0:-1]
        return oberste
    def __getList__(self):
        return self.st
    

class edge:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def __repr__(self):
        return f"Kante {self.a}|{self.b}"



#Aufgabe 7.0 b)

class graph:
    def __init__(self):
        self.kDict={} 
        self.edgeList=[]     #speichert Knotennamen und Adjazenten

    def addNode(self,k): #give name of graph and knot name
        if k in self.kDict:
            print(f"Knot {k} already exists")
        else: 
            self.kDict[k]=Stack() 
            print(f"knot {k} successfully created")

    def addEdge(self,k,j): #name of graph, a and b are knots you want to connect
        if (k and j) in self.kDict:
            self.kDict[k].__push__(j)
            self.kDict[j].__push__(k) #add both knots to their stacks
            #now creat a edge and add it to the graphs edgeList
            newEdge = edge(k,j)
            self.edgeList.append(newEdge)
            print(f"edge {newEdge} succesfully created")
        elif k in self.kDict:
            print(f"{j} is not a knot")
        elif j in self.kDict:
            print(f"{k} is not a knot")
        
#Aufgabe 7.0 c)
    def V(self):
        return list(self.kDict.keys())

#Aufgabe 7.0 d)
    def E(self):
        return self.edgeList

#Aufgabe 7.0 e)
    def allSingles(self):
        temp=[]
        for x in self.kDict: #sollte der Value eines Knotens eine leere Liste haben, existieren keine Kanten
            if self.kDict[x].__getList__()==[]:
                temp.append(x)
        return f"Die Knoten {temp} besitzen keine Kanten und/oder Nachbarn"

#Aufgabe 7.0 f)
    def mostEdges(self):
        meiste=0
        hoechste=[]
        knotenZahl=0
        for i in self.kDict:
            if len(self.kDict[i].__getList__())>meiste:
                knotenZahl=len(self.kDict[i].__getList__())
                meiste=i
        for i in self.kDict:
            if len(self.kDict[i].__getList__())==meiste:
                hoechste.append(i)
        print(f"Die Knoten {hoechste} besitzen {knotenZahl} und somit am meisten Knoten")
        


#Aufgabe 7.0 g)
    def neighbours(self,v):
        return self.kDict[v].__getList__()
          
```
